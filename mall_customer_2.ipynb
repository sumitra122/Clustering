# **Hierarcial clustering**

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, [3,4]].values
```

```
import scipy.cluster.hierarchy as sch 
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendogram Method')
plt.xlabel('Number of cluster')
plt.ylabel('Eucliean Distance')
plt.show()
```

```
from sklearn.cluster import AgglomerativeClustering
ag = AgglomerativeClustering(n_clusters = 5, affinity='euclidean', linkage = 'ward')
y_ag = ag.fit_predict(x)
print(y_ag)
```

```
plt.scatter(x[y_ag == 0, 0], x[y_ag == 0,1],s=100,c='red',label='cluster 1')
plt.scatter(x[y_ag == 1, 0], x[y_ag == 1,1],s=100,c='blue',label='cluster 2')
plt.scatter(x[y_ag == 2, 0], x[y_ag == 2,1],s=100,c='green',label='cluster 3')
plt.scatter(x[y_ag == 3, 0], x[y_ag == 3,1],s=100,c='cyan',label='cluster 4')
plt.scatter(x[y_ag == 4, 0], x[y_ag == 4,1],s=100,c='black',label='cluster 5')
plt.title('Clusters customer')
plt.legend()
plt.show()
```
